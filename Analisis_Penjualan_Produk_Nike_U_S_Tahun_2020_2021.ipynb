## **Datasets**

Dataset Penjualan Nike mencakup 9.360 transaksi penjualan di berbagai wilayah AS mulai dari 1 Januari 2020 hingga 31 Desember 2021.

Dataset diambil dari [Kaggle.com](https://www.kaggle.com/datasets/krishnavamsis/nike-sales) yang bersifat open source. Dataset tersebut mengandung 9 kolom, yaitu:

1. `Invoice Date`: Ini adalah tanggal ketika transaksi penjualan dilakukan atau tanggal faktur diterbitkan. Ini memberikan informasi tentang kapan penjualan dilakukan.

2. `Product` : Ini adalah nama atau kode produk yang terjual dalam transaksi tersebut. Ini memberikan informasi tentang produk yang dibeli oleh pelanggan.

3. `Region` : Ini adalah wilayah geografis di mana transaksi penjualan dilakukan. Misalnya, wilayah ini bisa merujuk pada negara bagian di AS atau wilayah lain yang relevan.

4. `Retailer` : Ini adalah nama pengecer atau toko tempat transaksi penjualan dilakukan. Ini memberikan informasi tentang dari mana produk dibeli.

5. `Sales Method` : Ini adalah metode penjualan yang digunakan dalam transaksi tersebut, seperti penjualan langsung, penjualan online, atau melalui distributor. Ini memberikan informasi tentang bagaimana produk dijual kepada pelanggan.

6. `State` : Ini adalah negara bagian di AS di mana transaksi penjualan terjadi. Ini bisa menjadi bagian penting dari analisis geografis.

7. `Price per Unit` : Ini adalah harga per unit atau harga individu dari produk yang terjual. Ini memberikan informasi tentang harga produk.

8. `Total Sales` : Ini adalah total pendapatan yang dihasilkan dari transaksi penjualan tersebut. Ini adalah hasil dari mengalikan harga per unit dengan jumlah unit yang terjual.

9. `Units Sold` : Ini adalah jumlah unit produk yang terjual dalam transaksi tersebut. Ini memberikan informasi tentang volume penjualan produk.

## Case 1: Melihat trend penjualan di tahun 2020 - 2021 setiap bulannya?

```python
# Membuat kolom baru untuk tahun dan bulan
listings['Year'] = listings['Invoice Date'].dt.year
listings['Month'] = listings['Invoice Date'].dt.month

# Filter data untuk tahun 2020 dan 2021
listings_2020 = listings[listings['Year'] == 2020]
listings_2021 = listings[listings['Year'] == 2021]

# Menghitung total penjualan per bulan untuk tahun 2020 dan 2021
total_sales_2020 = listings_2020.groupby('Month')['Total Sales'].mean()
total_sales_2021 = listings_2021.groupby('Month')['Total Sales'].mean()

# Membuat line chart untuk trend penjualan per bulan
plt.figure(figsize=(10, 6))
plt.plot(total_sales_2020.index, total_sales_2020.values, label='2020', marker='o')
plt.plot(total_sales_2021.index, total_sales_2021.values, label='2021', marker='o')

# Menambahkan judul
plt.title('Trend Penjualan per Bulan (2020 vs 2021)')

# Menambahkan label tiap axis
plt.xlabel('Bulan')
plt.ylabel('Rata-Rata Penjualan')

# Menambahkan legend
plt.legend()
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des'])

# tampilkan visualisasi
plt.show()
```

```python
# Kelompokkan data berdasarkan tahun dan hitung total penjualan
total_sales_by_year = listings.groupby('Year')['Total Sales'].sum()

# Menghitung persentase penjualan
sales_2020 = total_sales_by_year.loc[2020]
sales_2021 = total_sales_by_year.loc[2021]
percentage_2020 = (sales_2020 / (sales_2020 + sales_2021)) * 100
percentage_2021 = (sales_2021 / (sales_2020 + sales_2021)) * 100

# Plotting bar chart
plt.figure(figsize=(8, 6))
plt.bar(['2020', '2021'], [percentage_2020, percentage_2021], color=['blue', 'orange'])
plt.xlabel('Tahun')
plt.ylabel('Persentase Penjualan (%)')
plt.title('Persentase Penjualan antara Tahun 2020 dan 2021')
plt.ylim(0, 100)

# Menambahkan label pada tiap bar
for i, v in enumerate([percentage_2020, percentage_2021]):
    plt.text(i, v + 2, f'{v:.2f}%', ha='center')

plt.show()
```

Penjelasan:

* Syarat visualisasi ini ada dipastikan kolom `Invoice Date` sudah menjadi tipe `datetime`. Hal ini sudah diatasi di proses data wrangling
* Kemudian membuat kolom baru untuk tahun dan bulan dari tanggal transaksi.
* Melakukan filtering data untuk tahun 2020 dan 2021.
* Menghitung total penjualan per bulan untuk tahun 2020 dan 2021 menggunakan groupby dan sum.
* `marker='o'` digunakan untuk menambahkan titik-titik di setiap bulan pada line chart.
* Terakhir, membuat line chart menggunakan plt.plot() untuk membandingkan trend penjualan per bulan antara tahun 2020 dan 2021.

Analisis Bisnis:

```python
# Membuat kolom baru untuk tahun dan bulan
listings['Year'] = listings['Invoice Date'].dt.year
listings['Month'] = listings['Invoice Date'].dt.month

# Filter data untuk tahun 2020 dan 2021
listings_2020 = listings[listings['Year'] == 2020]
listings_2021 = listings[listings['Year'] == 2021]

# Hitung jumlah penjualan setiap produk untuk masing-masing tahun
sales_2020 = listings_2020.groupby('Product')['Units Sold'].mean().reset_index()
sales_2021 = listings_2021.groupby('Product')['Units Sold'].mean().reset_index()

# Pilih produk teratas untuk masing-masing tahun
top_product_2020 = sales_2020.sort_values(by='Units Sold', ascending=False)
top_product_2021 = sales_2021.sort_values(by='Units Sold', ascending=False)

# Buat visualisasi untuk produk teratas pada tahun 2020 dan 2021
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.bar(top_product_2020['Product'], top_product_2020['Units Sold'], color='skyblue')
plt.title('Produk Terlaris Tahun 2020')
# plt.xlabel('Produk')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
plt.bar(top_product_2021['Product'], top_product_2021['Units Sold'], color='lightgreen')
plt.title('Produk Terlaris Tahun 2021')
# plt.xlabel('Produk')
plt.ylabel('Jumlah Penjualan')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Melihat Korelasi Outlier Total Sales')

# Scatter plot: Units Sold vs Total Sales
sns.scatterplot(data=listings, x='Units Sold', y='Total Sales', ax=axes[0])
axes[0].set_title('Scatter Plot: Units Sold vs Total Sales')

# Scatter plot: Price per Unit vs Total Sales
sns.scatterplot(data=listings, x='Price per Unit', y='Total Sales', ax=axes[1])
axes[1].set_title('Scatter Plot: Price per Unit vs Total Sales')

# Box plot: Region vs Total Sales
sns.boxplot(data=listings, x='Region', y='Total Sales', ax=axes[2])
axes[2].set_title('Box Plot: Region vs Total Sales')

plt.tight_layout()
plt.show()
```

```python
# Hitung jumlah penjualan setiap produk
sales = listings.groupby('Product')['Units Sold'].mean().reset_index()

# Pilih produk teratas
top_product = sales.sort_values(by='Units Sold', ascending=False)

# Buat visualisasi untuk produk teratas pada tahun 2020 dan 2021
plt.figure(figsize=(10, 6))

plt.bar(top_product['Product'], top_product['Units Sold'], color='skyblue')
plt.title('Produk Terlaris Tahun 2020 - 2021')
plt.xlabel('Produk')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Melihat 5 produk Nike yang paling laris terjual (menggunakan barchart)
plt.figure(figsize=(10, 6))
top_products = listings['Product'].value_counts().nlargest(6)  # Mengambil 5 produk teratas
top_products.plot(kind='bar', color='skyblue')

plt.title('Top 5 Produk Nike Terlaris (2020 - 2021)')
plt.xlabel('Produk')
plt.ylabel('Jumlah Terjual')
plt.xticks(rotation=45, ha='right')

# Menambahkan jumlah angka di setiap batang
for i, count in enumerate(top_products):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

Penjelasan:

* **Mengambil 5 produk teratas:** Kita menggunakan metode value_counts() pada kolom "Product" untuk menghitung jumlah penjualan setiap produk, lalu memilih 5 produk teratas dengan metode nlargest(5).

* **Visualisasi barchart**
  * Membuat sebuah figure dengan ukuran 10x6 menggunakan `plt.figure(figsize=(10, 6))`.
  * Memberi judul pada barchart dengan `plt.title('Top 5 Produk Nike Terlaris')`, label sumbu x dengan `plt.xlabel('Produk')`, dan label sumbu y dengan `plt.ylabel('Jumlah Terjual')`.
